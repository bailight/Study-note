```python
import shutil
import os

def custom_copy_tree(src, dst):
    if os.path.exists(dst):
        shutil.rmtree(dst)
    return shutil.copytree(src, dst)

if not hasattr(shutil, 'copy_tree'):
    shutil.copy_tree = custom_copy_tree

import moabb
moabb.set_download_dir("D:\\moabb_datasets")

import moabb.datasets

m_dataset = moabb.datasets.bi2013a(
    NonAdaptive=True,
    Adaptive=True,
    Training=True,
    Online=True,
)

m_dataset.download()
```
